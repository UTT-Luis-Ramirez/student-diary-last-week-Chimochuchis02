##2. Medidas de Tendencia Central.
2.1. Media aritmetica(Promedio)
Para un conjunto dado de numeros X1, X2,....,Xn, la medida mas conocida y util del centro es la media aritmetica o promedio del conjunto. Esta se denotara por:  $\overline{x}$

$\overline{x}$ = $\frac{X1 + X2 +....+ Xn}{n} = \frac{1}{n}\sum_{i = 1}^{n} xi$

```python
#Cargamos el Dataset#
import pandas as pd
data = pd.read_csv('SoftwareProject_DataSet.csv')

#Creacion de una muestra aleatoria#

sample = data.sample(n=100, random_state=42)

#Verificacion del tamaño de la muestra#

print("La muestra tiene", len(sample), "observaciones")
sample.head()
```

```python
#Seleccion del valor que se pretende trabajar

code_lines_of_sample = sample["Líneas de código"]

#Creacion de la Funcion que realizara la operacion

def mean_calculator(data_list):
    sum_of_values = sum(data_list)
    sample_size = len(data_list)
    mean = sum_of_values / sample_size
    return mean

code_lines_mean = mean_calculator(code_lines_of_sample)

print("La media de los códigos de la muestra es:", code_lines_mean)
```

#2.2 La Mediana

**La palabra mediana es sinonimo de "medio y la mediana muestral es en realidad el valor medio una vez que se ordenan las observaciones de la mas pequeña a la mas grande. Cuando las observaciones estan denotadas por
 se utiliza el simbolo para representar la mediana muestral**

 Cuando el valor n es impar:

 x = x(n+1)/2

 Cuando el valor de n es par:

 x = 1/2(Xn/2 + X(x/2) + 1)

```python
#Creacion de la funcion que realizara la operacion

def median_calculator(data_list):
    sorted_list = sorted(data_list)
    sample_size = len(data_list)
    halfed_sample_size = sample_size // 2

    if sample_size % 2 == 0:
        return (sorted_list[halfed_sample_size-1] + sorted_list[halfed_sample_size]) / 2
    else:
        return sorted_list[halfed_sample_size]

code_line_median = median_calculator(code_lines_of_sample)

print(f"Mediana manual: {code_line_median} vs Pandas: {code_lines_of_sample.median()}")
```

#2.3 La Media Recortada

La media es bastante sensible a un solo valor extremo, mientras que la mediana es insensible a muchos valores apartados.

Una Muestra Recortada es un termino medio entre media
 y mediana. Una media 10% recortada, por ejemplo, se calcularia eliminando el 10% mas pequeño y el 10% mas grande de la muestra y luego promediando el resultante.

```python
def sample_cutter(sample, cut_percentage):
    sorted_sample = sorted(sample)
    sample_length = len(sample)
    index_to_cut = (cut_percentage / 100) * sample_length

    elements_to_cut = int(index_to_cut) if index_to_cut.is_integer() else int(index_to_cut) + 1

    sample_cutted = sorted_sample[elements_to_cut : sample_length - elements_to_cut]

    print("el tamaño de la muestra es: ", len(sample_cutted))

    return sample_cutted

    cutted_mean = mean_calculator(sample_cutter(code_lines_of_sample, 10))

print("La Media recortada es: ", cutted_mean)
```

#2.4 Cuartiles

Los cuartiles son los valores que dividen un conjunto de datos en cuatro partes iguales. Son una herramienta que tiene la estadistica para evaluar la tendencia central, la dispersión y la variabilidad de las observaciones

#2.5 Percentiles

Estos son valores que dividen un conjunto de datos en partes iguales y que indican el porcentaje de datos que son menores o igual a ellos. Los percentiles son una medida de posicion que se utiliza para coprar y contextualizar datos. Estos se representan con la letra
 P seguida por el número del percentil.

#3. Medidas de variabilidad

El reporte de una medida de centro da solo informacion parcial sobre un conjunto o distribucion de datos. Diferentes muestras o poblaciones pueden tener medidas identicas de centro y aun asi diferir entre si en otras importantes manera.

#3.1 Rango

El rango (tambien conocido como recorrido o amplitud) es una medida de dispersion que indica la diferencia entre el valor maximo y el valor minimo de un conjunto de datos. Es una forma sencilla de entender cuanto se extienden los datos.
